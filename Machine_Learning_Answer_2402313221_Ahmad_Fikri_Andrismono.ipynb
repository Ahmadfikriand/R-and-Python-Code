#### **1. Mean Squared Error (MSE) dan Mean Absolute Error (MAE) adalah dua metrik evaluasi yang sering digunakan untuk mengukur kualitas model regresi. Keduanya digunakan untuk menilai seberapa baik prediksi model sesuai dengan data sebenarnya.**
Mean Squared Error (MSE) adalah rata-rata dari kuadrat perbedaan antara nilai prediksi dan nilai sebenarnya. Dengan kata lain, setiap perbedaan (error) antara nilai sebenarnya dan prediksi dikuadratkan, dan kemudian dirata-ratakan.

MSE1.png

Mean Absolute Error (MAE) adalah rata-rata dari nilai absolut dari perbedaan antara nilai prediksi dan nilai sebenarnya. Dengan kata lain, kita hanya mengambil selisih mutlak antara prediksi dan nilai asli dan mengambil rata-ratanya.

MAE1.png

Berikut penjelasan dari rumus di atas:
- n adalah jumlah observasi
- xi adalah nilai sebenarnya untuk observasi ke-i
- yi adalah nilai prediksi untuk observasi ke-i

Kegunaan dalam Evaluasi Model Regresi:

#### MSE
Karena MSE mengkuadratkan kesalahan, ia akan memberikan bobot lebih pada kesalahan yang besar. Ini berarti model yang menghasilkan kesalahan besar akan dihukum lebih keras daripada model yang membuat kesalahan kecil. Dalam beberapa kasus, terutama ketika ingin menghindari kesalahan yang besar, MSE dapat menjadi metrik yang berguna.

#### MAE
MAE memberikan pandangan yang lebih langsung tentang kesalahan rata-rata. Kesalahan besar dan kecil dihukum sama. Jika ingin model yang lebih stabil dan tidak terlalu dipengaruhi oleh outlier, MAE mungkin lebih sesuai.

#### **4. Lakukan clustering data dalam 2 cluster menggunakan Gaussian Mixture Models (GMM).**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
```

```python
data = pd.read_excel('/content/clustering_gmm_dataset.xlsx')
data
```

```python
plt.scatter(data['x1'], data['x2'])
plt.show()
```

**a. Lakukan clustering data tersebut dalam 2 cluster menggunakan Gaussian Mixture Models (GMM)**

```python
# Algoritma Gaussian Mixture Models (GMM)
k = 2
gmm = GaussianMixture(n_components=k, random_state=42)
gmm.fit(data)
```

```python
# Parameter GMM setelah pelatihan selesai
means = gmm.means_
covariances = gmm.covariances_
weights = gmm.weights_
# Probabilitas keanggotaan untuk setiap cluster
probs = gmm.predict_proba(data)
```

```python
print('Means: ',means)
print('Covariances: ',covariances)
print('Weights: ',weights)
print('Probs: ',probs)
```

```python
# Tentukan cluster untuk setiap data berdasarkan probabilitas tertinggi
clusters = np.argmax(probs, axis=1)
clusters
```

```python
print("Parameter GMM:")
for i in range(k):
    print(f"Cluster {i}:")
    print("Mean:", means[i])
    print("Covariance:", covariances[i])
    print("Weight:", weights[i])
    print()
```

```python
print("Cluster untuk setiap data:")
for i, cluster in enumerate(clusters):
    print(f"Tanaman {i+1}: Cluster {cluster}")
```

```python
features = ['x1','x2']
X = data[features]
```

**b. Evaluasi dan Analisis**

```python
# Membuat dan mengatur GMM
gmm = GaussianMixture(n_components=2)  # misalnya, kita ingin 2 kluster
gmm.fit(X)

# Menghitung probabilitas untuk setiap kluster
probs = gmm.predict_proba(X)

# Mengambil kluster dengan probabilitas tertinggi
clusters = probs.argmax(axis=1)

# Mengatur syarat probabilitas > 0.5
clusters[probs.max(axis=1) > 0.5]
print(clusters)
```

```python
print("Cluster untuk setiap data:")
for i, cluster in enumerate(clusters):
    print(f"Tanaman {i+1}: Cluster {cluster}")
```
